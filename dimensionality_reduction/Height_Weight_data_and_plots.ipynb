```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
font = {'family':'sans-serif'}
```

```python
import random
```

```python
datapoints=100

x=[]
for i in range(datapoints):
    n=round(random.uniform(-2, 2),2)
    x.append(n)
```

```python
y=[]
for i in range(datapoints):
    n=round(random.uniform(x[i]-0.6, x[i]+0.6),2)
    y.append(n)
```

```python
plt.plot(x,y,'o',c='k')
plt.ylim(-3,3)
plt.xlim(-3,3)
plt.xlabel("Height",fontsize=16)
plt.ylabel("Weight",fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.tight_layout()

plt.savefig("HW01.png",dpi=1000)
```

```python
X=[]
Y=[]

for i in range(len(x)):
    X.append(x[i]+y[i])
    Y.append(x[i]-y[i])
```

```python
plt.plot(X,Y,'o',c='k')
plt.ylim(-5.2,5.2)
plt.xlim(-5.2,5.2)
plt.xlabel("Overall size = Weight + Height",fontsize=16)
plt.ylabel("Heavyness = Weight - Height",fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.tight_layout()

plt.savefig("HW02.png",dpi=1000)
```
