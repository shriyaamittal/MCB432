```python
import numpy as np
```

```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
## Date: March 7, 2019
```

```python
scores=np.loadtxt("data-01.txt")
```

```python
bins=[65,70,75,80,85,90,95,100,105,110]

plt.hist(scores, bins, histtype='bar', rwidth=0.8)
plt.title("Spring 2019 MCB 432 Assignment Scores",fontsize=14)
plt.ylabel("Students",fontsize=14)
plt.xlabel("Score",fontsize=14)
plt.yticks([0,1,2,3,4,5,6,7], fontsize=14)
plt.xticks([65,70,75,80,85,90,95,100,105,110], fontsize=14)

plt.savefig("hist-01.png",dpi=300)
```
